```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
cmap = {0 : [255  , 0    , 0    ],
        1 : [255  , 127.5, 0    ],
        2 : [255  , 225  , 0    ],
        3 : [127.5, 225  , 0    ],
        4 : [0    , 255  , 0    ],
        5 : [0    , 255  , 127.5],
        6 : [0    , 255  , 255  ],
        7 : [0    , 127.5, 255  ],
        8 : [0    , 0    , 255  ],
        9 : [127.5, 0    , 255  ],
        10: [255  , 0    , 255  ],
        11: [255  , 0    , 127.5]}

plt.figure(figsize=[12,4])

for i, value in enumerate(cmap.values()):
    plt.subplot(2, 6, i + 1)
    norm = np.array(value) / 255
    plt.imshow([[norm]])
    plt.title(value)
    plt.axis('off')
```
